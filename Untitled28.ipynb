```python
from pathlib import Path
from transformers import GPT2TokenizerFast

# Rutas a las carpetas
path_to_ballad = Path('/content/data/forms/ballad')
path_to_epistrophe = Path('/content/data/forms/epistrophe')
path_to_depression = Path('/content/data/topics/depression')
path_to_justice = Path('/content/data/topics/justice')

# Obtener la lista de archivos de texto en cada carpeta
ballad_files = list(path_to_ballad.glob('*.txt'))
epistrophe_files = list(path_to_epistrophe.glob('*.txt'))
depression_files = list(path_to_depression.glob('*.txt'))
justice_files = list(path_to_justice.glob('*.txt'))

# Cargar y tokenizar textos de 'ballad'
ballad_texts = []
for file_path in ballad_files:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        ballad_texts.append(text)

# Cargar y tokenizar textos de 'epistrophe'
epistrophe_texts = []
for file_path in epistrophe_files:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        epistrophe_texts.append(text)

# Cargar y tokenizar textos de 'depression'
depression_texts = []
for file_path in depression_files:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        depression_texts.append(text)

# Cargar y tokenizar textos de 'justice'
justice_texts = []
for file_path in justice_files:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        justice_texts.append(text)

# Tokenizar los textos
tokenizer = GPT2TokenizerFast.from_pretrained('gpt2')
tokenized_ballad_texts = [tokenizer.encode(text, return_tensors='pt') for text in ballad_texts]
tokenized_epistrophe_texts = [tokenizer.encode(text, return_tensors='pt') for text in epistrophe_texts]
tokenized_depression_texts = [tokenizer.encode(text, return_tensors='pt') for text in depression_texts]
tokenized_justice_texts = [tokenizer.encode(text, return_tensors='pt') for text in justice_texts]
```

```python
# Codifica el prompt en tokens usando el tokenizer
prompt_text = "Plato had an idea, he wanted to create a world where the gods did not have to sacrifice their lives to end life or even their children. I could have gone all the way. But he didn't know the cosmos."
prompt_input_ids = tokenizer.encode(prompt_text, return_tensors="pt")

# Genera texto con el modelo, asegurándote de pasar attention_mask
generated_text = model.generate(prompt_input_ids, max_length=500, temperature=0.9, pad_token_id=tokenizer.eos_token_id, do_sample=True)

# Decodifica y muestra el texto generado
decoded_text = tokenizer.decode(generated_text[0], skip_special_tokens=True)
print("Texto generado:")
print(decoded_text)


```
